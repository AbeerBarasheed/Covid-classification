load the libraries needed to build the CNN

```
from tensorflow.keras import datasets, models, layers, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from google.colab.patches import cv2_imshow
```

build the model and the layers then look at the model summary

```
# Set the CNN model

batch_size = None

model = models.Sequential()
model.add(layers.Conv2D(128, (5, 5), padding='same', activation="relu", input_shape=input_shape))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.2))
model.add(layers.Conv2D(64, (5, 5), padding='same', activation="relu"))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.3))
#model.add(layers.Conv2D(64, (3, 3), padding='same', activation="relu"))
#model.add(layers.BatchNormalization())
#model.add(layers.MaxPooling2D((2, 2)))
#model.add(layers.Dropout(0.4))
#model.add(layers.Conv2D(64, (3, 3), padding='same', activation="relu"))
#model.add(layers.BatchNormalization())
#model.add(layers.MaxPooling2D((2, 2)))
#model.add(layers.Dropout(0.5))

model.add(layers.GlobalMaxPooling2D())
model.add(layers.Dense(256, activation="relu"))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(3, activation="softmax"))

model.summary()
```

specify the optimizer

```
# initiate Adam optimizr
opt = optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08)
```

compile the model and use cross entropy as a loss function

```
# Let's train the model using RMSprop
model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
```

get the summary of the model again

```
# Network structure is summarized which confirms our design was implemented correctly.
model.summary()
```

Add the realystopping and the modelchecpoint

```
#Adding Early stopping callback to the fit function is going to stop the training,
#if the val_loss is not going to change even '0.001' for more than 10 continous epochs
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', min_delta=0.001, patience=10)

#Adding Model Checkpoint callback to the fit function is going to save the weights whenever val_loss achieves a new low value. 
#Hence saving the best weights occurred during training

model_checkpoint =  ModelCheckpoint('cifar_cnn_checkpoint_{epoch:02d}_loss{val_loss:.4f}.h5',
                                                           monitor='val_loss',
                                                           verbose=1,
                                                           save_best_only=True,
                                                           save_weights_only=True,
                                                           mode='auto',
                                                           period=1)
```

choose the batch size, and the epochs
Then fit the train_data and the train labels to the model call it history

```
# Set the batch size, number of epochs.
batch_size = 32
num_classes = 3
epochs = 40
num_predictions = 20

print('Hi')

history = model.fit(train_data, train_labels, batch_size=batch_size, epochs=epochs, verbose=1, 
                   validation_data=(test_data, test_labels),callbacks=[early_stopping,model_checkpoint])
#model.evaluate(test_data, test_labels)
```

plot the loss history of the train_data and the loss of the test data

```
# plot training history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
